# Most Popular Baby Names in California

```python
id_babies_2014 = states_babies_2014.loc[states_babies_2014['State']=='ID',:]
```

```python
sorted_id_babies = id_babies_2014.sort_values('Count', ascending = False)
```

```python
sorted_id_babies
```

```python
sorted_id_babies[0:5]
```

```python
id_babies = states_babies.loc[states_babies['State']=='ID', :]
```

```python
id_babies
```

```python
def popular(s):
    """Receives s, A Pandas Series, containing baby names in order of highest count to lowest count.
    Returns the most popular name in s."""
    return s.iloc[0]
```

```python
id_grouped = id_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)
```

```python
id_grouped
```
